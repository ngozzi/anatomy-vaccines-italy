```python
import pandas as pd

url = "https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv"

# Italian regions
north_west_regions = ["Piemonte", "Valle d'Aosta", "Liguria", "Lombardia"]
north_east_regions = ["P.A. Bolzano", 'P.A. Trento', "Veneto", "Friuli Venezia Giulia", "Emilia-Romagna"]
center_regions = ["Toscana", "Umbria", "Marche", "Lazio"]
south_regions = ["Abruzzo", "Molise", "Campania", "Puglia", "Basilicata", "Calabria"]
sicily_regions = ["Sicilia"]
sardinia_regions = ["Sardegna"]

df = pd.read_csv(url)
df["data"] = pd.to_datetime(df["data"])
df["data"] = df["data"].dt.strftime('%Y-%m-%d')
df.head()
```

```python
for savename, regions in zip(["Italy-northwest", "Italy-northeast", 
                              "Italy-center", "Italy-south", "Italy-sicily", "Italy-sardinia"],
                             [north_west_regions, north_east_regions, 
                              center_regions, south_regions, 
                              sicily_regions, sardinia_regions]):
    
    df_region = df.loc[df.denominazione_regione.isin(regions)].groupby("data", as_index=False).sum()
    df_region["daily_Deaths"] = df_region["deceduti"].diff()
    df_region = df_region[["data", "daily_Deaths", "nuovi_positivi"]]
    df_region.columns = ["Date", "daily_Deaths", "daily_Confirmed"]
    
    df_region.to_csv("../../basins/" + savename + "/epidemiological-data/epi_data.csv", index=False)
```
