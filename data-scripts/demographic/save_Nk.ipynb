```python
import pandas as pd
import numpy as np
# 'Nord-ovest', 'Nord-est', 'Centro', 'Sud', 'Sardegna', 'Sicilia'

def get_Nk_region(region):
    
    # loc region
    df_region = df.loc[(df["Territorio"] == region) & (df["Sesso"] == "totale")]
    group, value = [], []

    group.append("0-9")
    ages = ["Y" + str(i) for i in range(10)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("10-19")
    ages = ["Y" + str(i) for i in np.arange(10, 20, 1)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("20-24")
    ages = ["Y" + str(i) for i in np.arange(20, 25, 1)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("25-29")
    ages = ["Y" + str(i) for i in np.arange(25, 30, 1)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("30-39")
    ages = ["Y" + str(i) for i in np.arange(30, 40, 1)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("40-49")
    ages = ["Y" + str(i) for i in np.arange(40, 50, 1)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("50-59")
    ages = ["Y" + str(i) for i in np.arange(50, 60, 1)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("60-69")
    ages = ["Y" + str(i) for i in np.arange(60, 70, 1)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("70-79")
    ages = ["Y" + str(i) for i in np.arange(70, 80, 1)]
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    group.append("80+")
    ages = ["Y" + str(i) for i in np.arange(80, 100, 1)] + ['Y_GE100']
    value.append(df_region.loc[df_region["ETA1"].isin(ages)]["Value"].sum())

    Nk = pd.DataFrame(data={"group": group, "value": value})
    return Nk


df = pd.read_csv("./DCIS_POPRES1_06052021101115291.csv")
df.head()
```
